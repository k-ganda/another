```python
import gymnasium as gym
import torch
import pandas as pd
import ale_py
from stable_baselines3 import DQN
from stable_baselines3.dqn.policies import CnnPolicy

# Define environment
env_name = "ALE/Breakout-v5"
env = gym.make(env_name, render_mode="rgb_array")

# Define device
device = "cuda" if torch.cuda.is_available() else "cpu"

# Define hyperparameters (experimenting with values)
hyperparams = {
    "learning_rate": 1e-4,
    "gamma": 0.99,
    "batch_size": 8,
    "epsilon_start": 1.0,
    "epsilon_end": 0.05,
    "exploration_fraction": 0.1,
    "replay_buffer_size": 30000,
    "target_update_interval": 1000,
}

# Create the DQN model
model = DQN(
    policy=CnnPolicy,
    env=env,
    learning_rate=hyperparams["learning_rate"],
    gamma=hyperparams["gamma"],
    batch_size=hyperparams["batch_size"],
    exploration_initial_eps=hyperparams["epsilon_start"],
    exploration_final_eps=hyperparams["epsilon_end"],
    exploration_fraction=hyperparams["exploration_fraction"],
    buffer_size=hyperparams["replay_buffer_size"],
    target_update_interval=hyperparams["target_update_interval"],
    verbose=1,
    tensorboard_log="./dqn_logs/",
    device=device
)

# Function to evaluate model performance
def evaluate_model(model, env, num_episodes=5):
    rewards = []
    for episode in range(num_episodes):
        obs, _ = env.reset()
        total_reward = 0
        done, truncated = False, False
        while not (done or truncated):
            action, _ = model.predict(obs, deterministic=True)
            obs, reward, done, truncated, _ = env.step(action)
            total_reward += reward
        rewards.append(total_reward)
    return sum(rewards) / num_episodes  # Average reward

# Training and logging results
results = []
for step in [50000, 75000, 100000]:  # Log at different training steps
    model.learn(total_timesteps=step)
    avg_reward = evaluate_model(model, env)
    results.append({**hyperparams, "training_steps": step, "average_reward": avg_reward})
    print(f"Trained for {step} steps - Avg Reward: {avg_reward:.2f}")

# Save results to CSV for documentation
df = pd.DataFrame(results)
df.to_csv("hyperparameter_tuning_results.csv", index=False)

# Save trained model
model.save("dqn.zip")

env.close()
```
